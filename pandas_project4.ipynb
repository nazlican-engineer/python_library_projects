```python
#install datasets library
#!pip install datasets

#import libraries
import pandas as pd
from datasets import load_dataset

#loading data
dataset=load_dataset('lukebarousse/data_jobs')
df=dataset['train'].to_pandas()

#data cleanup
df["job_posted_date"]=pd.to_datetime(df["job_posted_date"])
```

```python
us_jobs=df[df.job_country=="United States"]
```

```python
us_jobs
```

```python
us_jobs.groupby("job_title_short")[["salary_year_avg"]].min() #[["salary_year_avg"]] burada iki parantez kullanmak verilerimi çizgi ve sütun grafikte görmemi sağlar.
```

```python
us_jobs.groupby("job_title_short")["salary_year_avg"].min()
```

```python
us_jobs.groupby("job_title_short")[["salary_year_avg"]].agg(["median","min","max","count"]).sort_values(by="median") #[["salary_year_avg"]] burada çift parantez kullanmaz gruplanmış
#dataframe üzerinden tekrar dataframe yao demek buda hata verir
```

```python
us_jobs.groupby("job_title_short")["salary_year_avg"].agg(["median","min","max","count"]).sort_values(by="median")
```

```python
us_jobs.groupby("job_title_short")["salary_year_avg"].agg(["median","min","max","count"]).sort_values(by="median",ascending=False)#ascending false artan sırada yapma azalan sırada yap
```
