```python
import matplotlib.pyplot  as plt
import numpy as np
#matplotlib.pyplot  modülü, Python programlama dilinde veri görselleştirme (grafik çizimi) yapmak için kullanılır.
```

```python
x = np.arange(0, 5, 0.1) #o dan başlayıp 0.1 artarak bi numpy dizisi oluşturur ve bu grafiğin  x eksenidir
y = np.sin(x) #y eksenine karşılık gelen değerler
plt.plot(x, y) # bu değerlere uygun  çizgi grafikleri oluşturmak için
plt.show()#yapılan grafiği göstermek için
```

```python
plt.scatter(x,y)
plt.show()
```
